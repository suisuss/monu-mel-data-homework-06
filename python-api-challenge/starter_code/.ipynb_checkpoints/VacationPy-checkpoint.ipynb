```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
#Load the CSV file from WeatherPy project

cities_file_path = "../output_data/cities_data.csv"

# Import the csv file using pandas
cities_df = pd.read_csv(cities_file_path)

cities_df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Access maps with unique API key
gmaps.configure(api_key=g_key)
```

```python
# Get parameters for gmap
locations = cities_df[["Lat", "Lon"]]
humidity = cities_df["Humidity"].astype(float)

#Plot the heatmap
figure = gmaps.figure(map_type='HYBRID')

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity,
                                 dissipating=True, max_intensity=50, point_radius=2.5)

# Add the head layer to the figure
figure.add_layer(heat_layer)

# display figure
figure
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Create a new df to allow for weather criteria. Max Temp lower than 80 bu thigher than 70
# Wind speed less than 10 mph, zero cloudiness.
# ^^ I've converted these numbers to the metric standard

# Creating new df based off of the orginal
conditions_df = cities_df.copy()
conditions_df = conditions_df[
    (conditions_df['Max Temp'] >= 21.111) & 
    (conditions_df['Max Temp'] <= 26.666) & 
    (conditions_df['Wind Speed'] <= 16.06) & 
    (conditions_df['Cloudiness'] == 0)
]

# Drop nulls and reset index
conditions_df = conditions_df.dropna()
conditions_df = conditions_df.reset_index()

# Drop any additional columns from your data set
conditions_df_drop = conditions_df.drop(['index', 'Unnamed: 0'], axis=1)
conditions_df_drop
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Create a dataframe with the columns that will be used for the project / add an additonal column for the hotel names

# Creating new df with empty Hotel Name column
hotel_df = pd.DataFrame(conditions_df_drop, columns=["City", "Country", "Lat", "Lon", "Hotel Name"])

hotel_df
```

```python
# Params dictionary for the requests
params = {
    "radius": 5000,
    "types": "hotel",
    "keyword": "hotel",
    "key": g_key
}

# Iterate each row
for index, row in hotel_df.iterrows():
    # Get lat, lng for each row
    lat = row["Lat"]
    lng = row["Lon"]

    # Add 'location' to params
    params["location"] = f"{lat},{lng}"

    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # Make request
    name_address = requests.get(base_url, params=params)
    
    # Convert to json
    name_address = name_address.json()
    
    # Try update df hotel name. Go on otherwise <--- will leave value as NaN
    try:
        hotel_df.iloc[index, 4] = name_address['results'][0]['name']
    except IndexError:
        continue
        

hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lon"]]
```

```python
# Add marker layer ontop of heat map

# Setup new figure
fig = gmaps.figure()

# Create marker layer
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)

# Add marker layer
fig.add_layer(markers)

fig
```
